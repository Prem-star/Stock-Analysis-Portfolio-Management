## Building the Model 

```
# creating covariance table on stock return dataframe
df1_cov = return_data1.cov()
print('Covariance Matrix:')
print(df1_cov)
print('\n') # return

# create the average of each stock
# these are the objective function COEFFICIENTS!
df1_return = return_data1.mean()
print('Average Return:')
print(df1_return)
```

```
from pyomo.environ import *
m = ConcreteModel()
```

```
m.ADM= Var(within=NonNegativeReals, bounds=(0,1))
m.CAG= Var(within=NonNegativeReals, bounds=(0,1))
m.CHD= Var(within=NonNegativeReals, bounds=(0,1))
m.CL= Var(within=NonNegativeReals, bounds=(0,1))
m.CLX= Var(within=NonNegativeReals, bounds=(0,1))
m.COST= Var(within=NonNegativeReals, bounds=(0,1))
m.EL= Var(within=NonNegativeReals, bounds=(0,1))
m.MNST= Var(within=NonNegativeReals, bounds=(0,1))
m.PG= Var(within=NonNegativeReals, bounds=(0,1))
m.WMT= Var(within=NonNegativeReals, bounds=(0,1))
```

```
# declare objective
m.objective = Objective(expr =               
                        m.ADM*df1_return[0]+
                        m.CAG*df1_return[1]+
                        m.CHD*df1_return[2]+
                        m.CL*df1_return[3]+
                        m.CLX*df1_return[4]+
                        m.COST*df1_return[5]+
                        m.EL*df1_return[6]+
                        m.MNST*df1_return[7]+
                        m.PG*df1_return[8]+
                        m.WMT*df1_return[9],
                        sense=maximize) # yes, we want to MAX our returns!

# declare constraints
# 1. Sum of all porportions = 1
m.sum_proportions = Constraint(expr =  m.ADM+m.CAG+m.CHD+m.CL+m.CLX+m.COST+m.EL+m.MNST+m.PG+m.WMT == 1)

# 2. Minimum return should be 0.0015 or 0.15%
m.return_floor = Constraint(expr = m.objective >= 0.0015)
```

Please note our Minimum return floor is based on daily returns and not on Monthly basis. 
- If we invest 1M USD, our daily return will be 1500 USD which is pretty good.
- We defined our return floor as 0.0015 or 0.15% because our data is based on daily activity of the stock performance. 

```
# creating calculations table for calculate the risk
# 3. Max risk should be less than 0.0003
max_risk = 0.0003
def calc_risk():
  # this is a list of all of your investment allocations (proportion invested in each stock)
  # this is UPDATED as the model is cranking - see how we are referncing the decision variables?
  proportions = [m.ADM,m.CAG,m.CHD,m.CL,m.CLX,m.COST,m.EL,m.MNST,m.PG,m.WMT]
  risk_matrix = [] # an empty list
  
  # now we iterate across the covariance matrix
  # 'risk elements' are = cov(A,B)*prop(A)*prop(B)
  # since there are 5 stocks, there are 5*5 = 25 elements
  # sum these 25 elements up at the end! that's the risk.
  for index, col in enumerate(df1_cov.columns):
    i = 0
    risk = []
    for row in df1_cov[col]:
      risk.append(proportions[index]*row*proportions[i])
      i = i + 1
    risk_matrix.append(risk)
  risk_matrix = pd.DataFrame(risk_matrix)
  return risk_matrix.sum().sum() 

# defining risk limits. This is what psioptparam() does.
# risk_limits = [0.0001, 0.0003] # this is just two options

# this is a sequence of options - better!
risk_limits = np.arange(0.0001, 0.0003, 0.00001) # take tiny steps
risk_limits
```

Please note again our risk ceilings have changed 
- We defined our risk ceilings to be less than our returns and our highest risk taking capacity is 0.0003 or 0.03% on daily basis 
- Since our data is based on daily stock performance activity, our risk capacity will be relatively smaller on daily basis

```
# updating risk contraint for each limit and then solving the problem
param_analysis = {} # key=risk, value =stock allocations 
returns = {} # key=risk, value = return
for r in risk_limits:
  m.total_risk = Constraint(expr = calc_risk() <= r)
  # run solver
  result = SolverFactory('ipopt', executable=ipopt_executable).solve(m).write()
  # store our allocation proportions
  param_analysis[r] = [m.ADM(),m.CAG(),m.CHD(),m.CL(),m.CLX(),m.COST(),m.EL(),m.MNST(),m.PG(),m.WMT()]
  # store our returns
  returns[r] = m.ADM()*df1_return[0]+m.CAG()*df1_return[1]+m.CHD()*df1_return[2]+m.CL()*df1_return[3]+m.CLX()*df1_return[4]+m.COST()*df1_return[5]+m.EL()*df1_return[6]+m.MNST()*df1_return[7]+m.PG()*df1_return[8]+m.WMT()*df1_return[9]
```

```
# generating the dataframe for proportions of the portfolio for each risk limit
param_analysis = pd.DataFrame.from_dict(param_analysis, orient='index')
param_analysis.columns = ['ADM', 'CAG', 'CHD', 'CL', 'CLX', 'COST', 'EL', 'MNST', 'PG','WMT']
```

```
pa1=pd.DataFrame(param_analysis)
pa1=pa1.reset_index()
pa1=pa1*100
```

## Optimal Stock Allocation

```
colors = px.colors.qualitative.Plotly
fig = go.Figure() 
j=0
for i in stocks:
  fig.add_traces(go.Scatter(name=i,x = pa1['index'],y = pa1[i],hovertext=i,mode = 'lines',line=dict(color=colors[j])))
  j+=1
fig.update_xaxes(title_text='Risk Percentage(%)')
fig.update_yaxes(title_text='Stock Allocation Percentage(%)')
fig.update_layout(title='Optimal Stock Allocation for Different Risk Levels',
                  font=dict(family="Arial, monospace",size=18,color=colors[0]))
fig.show()
```

## Concrete Model

```
# creating covariance table on stock return dataframe
df1_cov = return_data1.cov()
print('Covariance Matrix:')
print(df1_cov)
print('\n') # return

# create the average of each stock
# these are the objective function COEFFICIENTS!
df1_return = return_data1.mean()
print('Average Return:')
print(df1_return)
```

```
# declare objective
m.objective = Objective(expr =               
                        m.ADM*df1_return[0]+
                        m.CAG*df1_return[1]+
                        m.CHD*df1_return[2]+
                        m.CL*df1_return[3]+
                        m.CLX*df1_return[4]+
                        m.COST*df1_return[5]+
                        m.EL*df1_return[6]+
                        m.MNST*df1_return[7]+
                        m.PG*df1_return[8]+
                        m.WMT*df1_return[9],
                        sense=maximize) # yes, we want to MAX our returns!

# declare constraints
# 1. Sum of all porportions = 1
m.sum_proportions = Constraint(expr =  m.ADM+m.CAG+m.CHD+m.CL+m.CLX+m.COST+m.EL+m.MNST+m.PG+m.WMT == 1)

# 2. Minimum return should be 0.015 or 1.5%
m.return_floor = Constraint(expr = m.objective >= 0.015)
```

```
# creating calculations table for calculate the risk
# 3. Max risk should be less than 0.0015
max_risk = 0.007
def calc_risk():
  # this is a list of all of your investment allocations (proportion invested in each stock)
  # this is UPDATED as the model is cranking - see how we are referncing the decision variables?
  proportions = [m.ADM,m.CAG,m.CHD,m.CL,m.CLX,m.COST,m.EL,m.MNST,m.PG,m.WMT]
  risk_matrix = [] # an empty list
  
  # now we iterate across the covariance matrix
  # 'risk elements' are = cov(A,B)*prop(A)*prop(B)
  # since there are 5 stocks, there are 5*5 = 25 elements
  # sum these 25 elements up at the end! that's the risk.
  for index, col in enumerate(df1_cov.columns):
    i = 0
    risk = []
    for row in df1_cov[col]:
      risk.append(proportions[index]*row*proportions[i])
      i = i + 1
    risk_matrix.append(risk)
  risk_matrix = pd.DataFrame(risk_matrix)
  return risk_matrix.sum().sum() 

# defining risk limits. This is what psioptparam() does.
# risk_limits = [0.001, 0.0015] # this is just two options

# this is a sequence of options - better!
risk_limits = np.arange(0.001, 0.007, 0.00005) # take tiny steps
risk_limits
```

```
# updating risk contraint for each limit and then solving the problem
param_analysis = {} # key=risk, value =stock allocations 
returns = {} # key=risk, value = return
for r in risk_limits:
  m.total_risk = Constraint(expr = calc_risk() <= r)
  # run solver
  result = SolverFactory('ipopt', executable=ipopt_executable).solve(m).write()
  # store our allocation proportions
  param_analysis[r] = [m.ADM(),m.CAG(),m.CHD(),m.CL(),m.CLX(),m.COST(),m.EL(),m.MNST(),m.PG(),m.WMT()]
  # store our returns
  returns[r] = m.ADM()*df1_return[0]+m.CAG()*df1_return[1]+m.CHD()*df1_return[2]+m.CL()*df1_return[3]+m.CLX()*df1_return[4]+m.COST()*df1_return[5]+m.EL()*df1_return[6]+m.MNST()*df1_return[7]+m.PG()*df1_return[8]+m.WMT()*df1_return[9]
```

```
# generating the dataframe for proportions of the portfolio for each risk limit
param_analysis = pd.DataFrame.from_dict(param_analysis, orient='index')
param_analysis.columns = ['ADM', 'CAG', 'CHD', 'CL', 'CLX', 'COST', 'EL', 'MNST', 'PG','WMT']
```
